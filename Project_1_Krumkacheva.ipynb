Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

```python
#ваш код здесь

# display (data.head())

column_edu_and_uni = 'Образование и ВУЗ'
column_edu = 'Образование'
# display (data[column_edu_and_uni].head())

# На выходе функции возмжоны четыре варианта:
# - высшее, 
# - неоконченное высшее, 
# - среднее специальное, 
# - среднее.
def parse_edu(obj_edu_and_uni : object):
    # Преобразуем в тип Строка
    str_edu_and_uni = str(obj_edu_and_uni)
    
    # Преобразуем буквы в нижний регистр, 
    # убираем лишние пробелы (в начале и в конце строки)
    str_edu_and_uni = str_edu_and_uni.lower().strip()
    
    # Оставим в строке только первые три слова. Для этого:
    # 1. Сначала разобъём строку на слова
    words = str_edu_and_uni.split()
    i = 0
    str_edu_and_uni = ''
    # 2. Возьмём перевые три слова либо меньше (если слов меньше трёх)
    # и склеим снова слова в строку
    for word in words:
        str_edu_and_uni += word
        i += 1
        if (i >= 3):
            break
    
    result = ''
    
    if (str_edu_and_uni.find('высшее') >= 0):
        # Значит образование 'высшее' либо 'неоконченное высшее'
        if (str_edu_and_uni.find('неоконченное') >= 0):
            result = 'неоконченное высшее'
        else:
            result = 'высшее'
    elif (str_edu_and_uni.find('среднее') >= 0):
        # Значит образование 'среднее' либо 'среднее специальное'
        if (str_edu_and_uni.find('специальное') >= 0):
            result = 'среднее специальное'
        else:
            result = 'среднее'
    else:
        # Значит ни одно из четырёх перечисленных выше
        print('Error')
        
    return result    

# Добавим новый столбец 'Образование'
# вычислив его на основе столбца 'Образование и ВУЗ'
data[column_edu] = data[column_edu_and_uni].apply(parse_edu)
# display(data[[column_edu, column_edu_and_uni]])

data_3_1 = data.groupby(by=column_edu).count()
# display(data_3_1.head())

# Оставим только вспомогательный столбец (в нём подсчитано кол-во строк)
data_3_1 = data_3_1[[column_edu_and_uni]]

# Переименовываем вспомогательный столбец в 'Count'
data_3_1 = data_3_1.rename(columns={column_edu_and_uni: 'Count'})

# Сортируем по столбцу 'Count' в порядке убывания
data_3_1 = data_3_1.sort_values(by=['Count'], ascending=False)

display (data_3_1.head())

# Выполните преобразование, ответьте на контрольные вопросы и удалите признак «Образование и ВУЗ».
display (data.head())
data = data.drop(columns=[column_edu_and_uni])
display (data.head())
```

```python
#ваш код здесь

column_work_exp    = 'Опыт работы'
column_work_months = 'Опыт работы (месяц)'

def express_analysis():
    i = 0
    num_good = 0
    num_nan = 0
    for work_exp in data[column_work_exp]:
        i += 1
        str_work_exp = str(work_exp).lower()
        if (str_work_exp.find('опыт работы ') >= 0):
            num_good += 1
        elif (str_work_exp == 'nan'):
            num_nan += 1
        else:
            num_nan += 1

    print ('Express Analysis:')
    print ('good = ', num_good, '/', i)
    print ('nan  = ', num_nan, '/', i)
    print ('good + nan =', num_good + num_nan, '/', i)
    print ()

def parse_work_exp(obj_work_exp : object):
    # Преобразуем в тип Строка
    str_work_exp = str(obj_work_exp)
    
    # Преобразуем буквы в нижний регистр, 
    # убираем лишние пробелы (в начале и в конце строки)
    str_work_exp = str_work_exp.lower().strip()
    
    # Вычисляем 'Опыт работы (месяц)'
    result_months = 0
    
    if (str_work_exp == ''):
        # Обработка случая, если строка пустая
        result_months = np.nan
    
    elif (str_work_exp == 'nan'):
        # Обработка случая, если строка равна Not a number
        result_months = np.nan
    
    elif (str_work_exp.find('опыт работы') >= 0):
        # Обработка наиболее ожидаемого (вероятного) случая - 
        # в строке есть подстрока 'опыт работы'
        # Тогда эту строку нужно исследовать дальше
        # Пример #1 (было): 'опыт работы 3 месяца  октябрь 2004 — декабрь ...'
        # Пример #2 (было): 'опыт работы 10 лет 3 месяца  октябрь 2004 — декабрь ...'
        
        # Укорачиваем строку слева на фразу 'опыт работы'
        # Пример #1 (станет): ' 3 месяца  октябрь 2004 — декабрь ...'
        # Пример #2 (станет): ' 10 лет 3 месяца  октябрь 2004 — декабрь ...'
        idx = str_work_exp.find('опыт работы')
        idx += len('опыт работы')
        str_work_exp = str_work_exp[idx:]
        
        # Разбиваем на слова
        # Далее нас будет интересовать максимум первые 4 слова
        # Пример #1 (станет): ['3', 'месяца', ...]
        # Пример #2 (станет): ['10', 'лет', '3', 'месяца', ...]
        words = str_work_exp.strip().split()
        
        if (len(words) >= 2):
            # Ожидаем, что кол-во слов не мене двух: 
            # - чтобы извлечь только месяцы (стаж <  1 года) или 
            # - чтобы извлечь годы и месяцы (стаж >= 1 года)             
                    
            try:
                # Первым по счёту должно идти число
                # Пример #1: words = ['3', 'месяца', ...], тогда num = 3
                # Пример #2: words = ['10', 'лет', '3', 'месяца', ...], тогда num = 10
                num = int(words[0].strip())
            except:
                # Обработка неожиданного случая:
                # - первым идёт не целое число
                # Потребуется дописать код.
                print ('Error: Work Experience #1')
                result_months = np.nan
                return result_months
            
            if (words[1].find('мес') >= 0):
                # Если следом за числом идёт, что число указано в месяцах
                # Тогда просто присваиваем его к результату
                # Пример #1: words = ['3', 'месяца', ...] 
                # Тогда result_months = 3 мес.
                result_months = num
                
            elif (words[1].find('лет') >= 0 or
                  words[1].find('год') >= 0):
                # Если следом за числом идёт, что число указано в годах
                # Тогда переводим годы в месяцы
                # Пример #2: words = ['10', 'лет', '3', 'месяца', ...]
                # Тогда result_months = 10 * 12 мес. = 120 мес.
                result_months = num * 12
                
                # Проверяем, что ещё есть слова
                if (len(words) >= 4):
                    # Далее могут идти ещё месяцы (а могут не идти)    
                    if (words[3].find('меся') >= 0):
                        # Если далее идут месяцы, то будем считывать число
                        
                        try:
                            num = int(words[2].strip())
                            result_months += num
                        except:
                            # Обработка неожиданного случая:
                            # - на позиции слова #3 идёт не целое число
                            # Потребуется дописать код.
                            print ('Error: Work Experience #2')
                            print (str_work_exp)
            
            else:
                # Обработка неожиданного случая (пока что необрабатываемого случая)
                print ('Error: Work Experience #3')
            
        else:
            # Обработка неожиданного случая (пока что необрабатываемого случая)
            print ('Error: Work Experience #4')
    
    else:
        result_months = np.nan
        
    return result_months

# display (data.info())

# Проведём экспресс-анализ, чтобы посмотреть:
# - кол-во "хороших" ячеек (которые можно парсить)
# - кол-во ячеек с NaN
express_analysis()

# Добавим новый столбец 'Опыт работы (месяц)'
# вычислив его на основе столбца 'Образование и ВУЗ'
data[column_work_months] = data[column_work_exp].apply(parse_work_exp)
display (data[column_work_months].info())

# Чему равен медианный опыт работы (в месяцах) в нашей таблице?
print ()
print ('mean   =', data[column_work_months].mean())
print ('median =', data[column_work_months].median())

# Выполните преобразование и удалите столбец «Опыт работы» из таблицы.
display (data.head())
data = data.drop(columns=[column_work_exp])
display (data.head())
```
